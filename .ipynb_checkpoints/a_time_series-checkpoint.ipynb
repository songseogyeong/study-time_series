### 시계열 데이터
- 시간이 지남에 따라 기록된 정보들을 의미하고, 관측시간에 대한 관측자료의 관계로 표현한다.
- 시간 흐름에 따라 변화하는 방향성을 이해하고, 그 변화를 예측하는 것이 주 목적이다.
- 여름 휴가 기간 동안 특정 해변 도시의 방문자 수를 기록하고 분석하면, 일정한 패턴이나 추세를 찾게 된다.  
  이를 통해 다음 해의 방문자 수를 예측하여 관광 산업이나 여행 업계에서 호율적인 운영 및 마케팅 전략을 계획할 수 있다.  
  <sub>*예를 들어, 양양 해변 도시 방문자를 매일 기록하면, 전체적인 패턴이 보인다. 여름에 많은 패턴이 있을 수 있고 점점 방문자 수가 늘어나면 오르는 추세이다.</sub>
- 시계열 데이터의 유형에는 우연변동 시계열, 계절 변동 시계열, 추세 변동 시계열, 계절-추세 변동 시계열 데이터가 있다.
- 자료의 인덱스가 시간인 데이터이다.  
  <sub>*index가 시간과 날짜인 데이터 프레임을 **시계열 데이터**(time series)라고 한다.</sub>   
  <sub>*시계열 데이터에서는 이상치(bias)를 쇼크라고 한다. 올해, 작년, 재작년 데이터를 모두 보고 쇼크까지 추가하여 예측하면 된다.</sub>

#### 우연변동 시계열 데이터
- 특정한 패턴이나 추세 없이 우연한 요소에 의해 변하는 데이터이다.
- 패턴이나 규칙이 없고 우연히 발생하는 것처럼 보인다.  
<sub>*패턴이 없고 랜덤하며, 우연성 시계열 데이터라고 부르기도 한다.    
*주식이 우연변동 시계열 데이터라고 볼 수 있다.</sub>

```python
import pandas as pd
import matplotlib.pyplot as plt

# 우연변동 시계열 데이터 생성 (패턴이 없고 랜덤)
random1 = [1342, 1442, 1252, 1343,
           1425, 1362, 1256, 1272,
           1243, 1359, 1412, 1253,
           1201, 1478, 1322, 1406,
           1254, 1289, 1497, 1208]

# date_range 라이브러리를 사용하여 날짜 또는 시간의 배열을 생성
# start = 시작 날짜, periods = 기간(정수로 전달), freq = 빈도/주기
# 날짜에는 주기가 있음(매일/월/년/분기), start 시 올해 날짜를 작성하면 미래의 것이 호출되기 때문에 올해가 호출될 수 있도록 19년도로 설정
# 데이터 수와 인덱스 수는 동일해야하므로 len을 사용하여 데이터 개수 그대로를 넣어주기
# freq: 'D' 기본값 D(Day)는 1일을 의미한다.
# '5D'일 경우 5일(Day) 단위이고 '2W'일 경우 2주(Week) 단위이다.
# '3M'일 경우 3개월(Month) 단위이고, 분기별로 나타낼 때에는 'Q(Quarter)'를 사용한다(3월, 6월, 9월, 12월).
idx = pd.date_range(start='2019-01', periods=len(random1), freq='Q')

# 시계열 데이터로 변환
random1_ts = pd.DataFrame(random1, index=idx)

# 데이터 프레임 마지막 5행 출력
# 인덱스가 날짜이며, 단변량 데이터다. (변량이 하나)
display(random1_ts.tail())

# 시각화
plt.figure(figsize=(10, 6))
for col in random1_ts.columns:
    plt.plot(random1_ts.index, random1_ts[col], label=col)

plt.title('Random Variation Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

### 계절변동 시계열 데이터
- 특정한 시간 주기(예: 일년)에 따라 발생하는 패턴이 나타나는 데이터이다.
- 특정 시간에 특정한 일이 반복된다.  
  <sub>*계절성 시계열 데이터라고 부르기도 한다.  
  *물결의 형태를 띄고 있다.</sub>

```python
import pandas as pd
import matplotlib.pyplot as plt

# 계절변동 데이터 생성
season1 = [1142, 1242, 1452, 1543,
           1125, 1262, 1456, 1572,
           1143, 1269, 1462, 1553,
           1121, 1258, 1472, 1546,
           1154, 1249, 1477, 1548]

# 시계열 데이터로 변환
idx = pd.date_range(start='2019-01', periods=len(season1), freq='Q')
season1_ts = pd.DataFrame(season1, index=idx)

display(season1_ts.tail())

# 시각화
plt.figure(figsize=(10, 6))
for col in season1_ts.columns:
    plt.plot(season1_ts.index, season1_ts[col], label=col)

plt.title('Random Variation Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# 그래프 확인 시 연초에 데이터가 상승하는것을 확인할 수 있다.
# 위 우연변동 시계열 데이터와 비교하였을 때 패턴이 드러나는 것을 확인할 수 있다. (계절성, 시즈널/seasonal)
```

### 추세변동 시계열 데이터
- 시간에 따라 일정한 방향으로 계속해서 증가하거나 감소하는 추세를 보이는 데이터이다.
- 일정한 방향으로 계속해서 증가하거나 감소한다.  
  <sub>*기업에서 매출액 분석 시 추세변동 시계열 데이터가 제일 많이 보인다.  
  *얼핏보면 우연변동 시계열 데이터로 보일 수 있으나 멀리서 보면 추세변동 시계열 데이터인 경우가 있다.  
  멀리봐야 보이는 데이터이며, 육안으로는 확인이 힘든 부분이 있어 자세히 확인해야 한다.</sub>

```python
import pandas as pd
import matplotlib.pyplot as plt

# 추세변동 시계열 데이터 생성
trend1 = [1142, 1242, 1252, 1343,
          1225, 1562, 1356, 1572,
          1343, 1459, 1412, 1453,
          1401, 1478, 1322, 1606,
          1554, 1589, 1597, 1408]

# 시계열 데이터로 변환
idx = pd.date_range(start='2019-01', periods=len(trend1), freq='Q')
trend1_ts = pd.DataFrame(trend1, index=idx)

display(trend1_ts.tail())

# 시각화
plt.figure(figsize=(10, 6))
for col in trend1_ts.columns:
    plt.plot(trend1_ts.index, trend1_ts[col], label=col)

plt.title('Random Variation Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# 아래 그래프는 오름 추세를 보이고 있다.
```

### 계절-추세변동 시계열 데이터
- 추세와 계절성 요인이 함계 나타나는 데이터로서, 시간에 따라 일정한 추세가 있고 특정 시간 주기에 따른 패턴이 있는 데이터이다.
- 시간이 지남에 따라 일정한 방향으로 움직이면서도 특정 시간에 특정한 일이 반복된다.  
  <sub>*특정한 시간에 따라 반복되는데 주기도 있는 것을 의미한다. 즉, 계절성도 있는데 추세도 있다는 것이다.  
  *플랫하면 단순히 계절변동이며, 주기가 없으면 추세 변동이다. 추세마저 없으면 우연변동 시계열 데이터다.</sub>

```python
import pandas as pd
import matplotlib.pyplot as plt

# 계절-추세변동 시계열 데이터 생성
st1 = [1142, 1242, 1452, 1543,
       1225, 1362, 1556, 1672,
       1343, 1459, 1662, 1753,
       1421, 1558, 1772, 1846,
       1554, 1649, 1877, 1948]

# 시계열 데이터로 변환
idx = pd.date_range(start='2019-01', periods=len(st1), freq='Q')
st1_ts = pd.DataFrame(st1, index=idx)

display(st1_ts.tail())

# 시각화
plt.figure(figsize=(10, 6))
for col in st1_ts.columns:
    plt.plot(st1_ts.index, st1_ts[col], label=col)

plt.title('Random Variation Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# 시계열 데이터 생성
random1_ts = pd.Series([1342, 1442, 1252, 1343, 1425, 1362, 1256, 1272,
                        1243, 1359, 1412, 1253, 1201, 1478, 1322, 1406,
                        1254, 1289, 1497, 1208],
                       index=pd.date_range(start='2019-01', periods=20, freq='Q'))

trend1_ts = pd.Series([1142, 1242, 1252, 1343, 1225, 1562, 1356, 1572,
                       1343, 1459, 1412, 1453, 1401, 1478, 1322, 1606,
                       1554, 1589, 1597, 1408],
                      index=pd.date_range(start='2019-01', periods=20, freq='Q'))

season1_ts = pd.Series([1142, 1242, 1452, 1543, 1125, 1262, 1456, 1572,
                        1143, 1269, 1462, 1553, 1121, 1258, 1472, 1546,
                        1154, 1249, 1477, 1548],
                       index=pd.date_range(start='2019-01', periods=20, freq='Q'))

st1_ts = pd.Series([1142, 1242, 1452, 1543, 1225, 1362, 1556, 1672,
                    1343, 1459, 1662, 1753, 1421, 1558, 1772, 1846,
                    1554, 1649, 1877, 1948],
                   index=pd.date_range(start='2019-01', periods=20, freq='Q'))

# 그래프 그리기
plt.figure(figsize=(14, 10))

# Random Variation Series
plt.subplot(2, 2, 1)
# resample: 시계열 데이터의 빈도를 변경하는 기능
# 일별을 주별, 월별, 분기별, 연간 으로 변경하여 큰 간격으로 집계하거나 요약 가능
# Q(분기) 별로 sum을하여 원형 마커로 표기
random1_ts.resample('Q').sum().plot(marker='o')
plt.title('EDA: Random Variation Series')
plt.xlabel('Quarter: 2019-2023')
plt.ylabel('Sales')

# Trend Variation Series
plt.subplot(2, 2, 2)
trend1_ts.resample('Q').sum().plot(marker='o')
plt.title('EDA: Trend Variation Series')
plt.xlabel('Quarter: 2019-2023')
plt.ylabel('Sales')

# Seasonal Variation Series
plt.subplot(2, 2, 3)
season1_ts.resample('Q').sum().plot(marker='o')
plt.title('EDA: Seasonal Variation Series')
plt.xlabel('Quarter: 2019-2023')
plt.ylabel('Sales')

# Seasonal-Trend Variation Series
plt.subplot(2, 2, 4)
st1_ts.resample('Q').sum().plot(marker='o')
plt.title('EDA: Seasonal-Trend Variation Series')
plt.xlabel('Quarter: 2019-2023')
plt.ylabel('Sales')

plt.tight_layout()
plt.show()
```
