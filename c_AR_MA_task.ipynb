시계열 데이터는 시작과 끝이 파일명에 있어야함!

```python
import pandas as pd
import json

# with문 쓰는 이유는 자동 close가 됙 ㅣ때문에 
with open('./datasets/bitcoin_20210_2024.json') as f:
    # 불러오 ㄹ때 자동으로 dict으로 나옴
    json_data = json.load(f)

# 필요한 세번째 데이터list만 가져오기
bitcoin_df = pd.DataFrame(json_data['market-price'])
bitcoin_df
```

```python
from datetime import datetime

def changeDate(milis):
    # 초로 변경
    timestamp = milis / 1000
    # 앞에 있는 초를 뒤 형식으로 변경해라!
    converted_time = datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d')
    return converted_time
```

```python
bitcoin_df.x = bitcoin_df.x.apply(changeDate)
bitcoin_df
```

```python
bitcoin_df.set_index('x', inplace=True)
bitcoin_df
```

```python
bitcoin_df.info()
```

```python
import pandas as pd

bitcoin_df.index = pd.to_datetime(bitcoin_df.index)
bitcoin_df.info()
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, ax = plt.subplots(1, 2, figsize=(12, 8))
plot_acf(bitcoin_df, lags=20, ax=ax[0])
plot_pacf(bitcoin_df.diff().dropna(), lags=20, ax=ax[1])

plt.show()
```

```python
# 차분
bitcoin_df = bitcoin_df_2017_2024.diff()
# NaN값 제거
bitcoin_df.dropna(inplace=True)
bitcoin_df
```

```python
window_size = 50

moving_avg = bitcoin_df.rolling(window_size).mean()
display(moving_avg)

moving_avg_diff = bitcoin_df.rolling(window_size).mean()
display(bitcoin_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))

plt.subplot(2, 1, 1)
plt.plot(bitcoin_df, c='orange')

plt.subplot(2, 1, 1)
plt.plot(moving_avg_diff)

plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(bitcoin_df, order=(1, 4, 21))
model_fit = model.fit()

start_index = pd.to_datetime('2017-07-20')
end_index = pd.to_datetime('2024-06-09')
forcast = model_fit.predict(start=start_index, end=end_index)
```

```python
plt.figure(figsize=(15, 8))
plt.plot(bitcoin_df, label='original')
plt.plot(forcast, label='predicted', c='orange')
plt.title("Time Series Forecast")
plt.legend()
plt.show()
```
